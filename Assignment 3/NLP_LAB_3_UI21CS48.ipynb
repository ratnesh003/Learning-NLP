```python
import string
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import csv
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/drive/MyDrive/ML/Job_skills.csv', encoding = "ISO-8859-1")
# df.pop('Unnamed: 2')
# df.pop('Unnamed: 3')
# df.pop('Unnamed: 4')
df.head(10)
```

```python
text = df['Preferred Qualifications']
n = len(df)
lowercased = [None]*n
for i in range(n):
  if isinstance(text[i], str):
    lowercased[i] = text[i].lower()
  else:
    lowercased[i] = text[i]
df['lowercased'] = lowercased
df.head(10)
```

```python
text = df['lowercased']
n = len(text)
punctuized = []
for i in range(n):
  if isinstance(text[i], str):
    punctuized.append(text[i].translate(str.maketrans('', '', string.punctuation)))
  else:
    punctuized.append(text[i])

df['punctuized'] = punctuized
df.head(10)
```

```python
text  = df['punctuized']
n = len(text)
tokenized = [None]*n
for i in range(n):
  # Check if the value is a string before applying split()
  if isinstance(text[i], str):
    tokenized[i] = text[i].split()
  # If not a string (e.g., float due to NaN), assign an empty list or handle as needed
  else:
    tokenized[i] = []  # or tokenized[i] = text[i] to keep the original value
df['tokenized'] = tokenized
df.head(10)
```

```python
text = df['tokenized']
n = len(text)
SWR = [None]*n
print("size = ", n)

sw = ['a', 'an', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 'such', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']
count = 0

for i in range(n):
  swrArr=[]
  for word in text[i]:
    if word in sw:
      continue
    else:
      swrArr.append(word)
  SWR[i] = swrArr

df['SWR'] = SWR
df.head()
```

```python
porter_stemmer = PorterStemmer()

text = df['SWR']
n = len(df)
Stemmed = [None]*n
print("size = ", n)

for i in range(n):
  stemmed=[]
  for word in text[i]:
    shabd = porter_stemmer.stem(word)
    stemmed.append(shabd)
  Stemmed[i] = stemmed

df['Stemmed'] = Stemmed
df.head()
```

```python
df.drop(columns=['Stemmed'], inplace=True)
df.head()
```

```python
text = df['SWR']
n = len(df)
lemmatized = [None]*n
print("size = ", n)

for i in range(n):
  lemmatized[i] = [porter_stemmer.stem(word) for word in text[i]]

df['lemmatized'] = lemmatized
df.head()
```

```python
import csv
with open('/content/drive/MyDrive/ML/word_analysis.csv', 'w', newline='') as file:
  writer = csv.writer(file)
  writer.writerow(['word', 'atomic', 'stemmed'])
```

```python
text = df['lemmatized']
n = len(text)
atomic = [None]*n
stemmed = [None]*n
words = [None]*n
with open('stemmed_words.csv', mode='w', newline='') as file:
  writer = csv.writer(file)
  for i in range(n):
    for word in text[i]:
      words[i] = word
      stemmed_word = porter_stemmer.stem(word)
      stemmed[i] = stemmed_word
      if(stemmed_word != word):
        atomic[i] = True
      else:
        atomic[i] = False

df2=pd.read_csv('/content/drive/MyDrive/ML/word_analysis.csv', encoding = "ISO-8859-1")
df2['words'] = words
df2['atomic'] = atomic
df2['stemmed'] = stemmed
df2.to_csv('/content/drive/MyDrive/ML/word_analysis.csv', index=False, encoding='ISO-8859-1')
df2.head(2000)
```
