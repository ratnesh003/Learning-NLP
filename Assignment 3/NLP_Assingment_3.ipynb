```python
import nltk
import pandas as pd
from nltk.corpus import words
import matplotlib.pyplot as plt
nltk.download('words')
```

```python
english_words = set(words.words())
```

```python
def generate_words(root_words, suffixes):
    meaningful_words = []
    data = []

    for root in root_words:
        row = {}
        for suffix in suffixes:
            word = root + suffix
            if word.lower() in english_words:
                meaningful_words.append(word)
                row[suffix] = word
            else:
                row[suffix] = False
        data.append(row)

    return meaningful_words, data
```

```python
root_words = [
    "act", "play", "move", "arrange", "hope", "run", "fix", "color", "test", "talk", "dance", "read",
    "write", "speak", "love", "care", "drive", "draw", "paint", "listen", "fight", "wait", "watch",
    "walk", "help", "jump", "teach", "learn", "cook", "clean", "fly", "build", "change", "carry",
    "sing", "swim", "study", "work", "give", "find", "hold", "think", "hear", "smile", "dream",
    "play", "trust", "guess", "plan"
]

suffixes = [
    "ing", "ed", "er", "ment", "ful", "s", "able", "less", "ly", "ness", "ation", "or", "al",
    "est", "ish", "ize", "tion", "ance", "ence", "ity", "ion", "ous", "ive", "ent",
    "ian", "ise", "ate", "ic", "ary", "ure", "ant", "eer", "ence", "ist", "ism", "ify",
    "ology", "ship", "dom", "hood", "ward", "wise", "wards", "y"
]

meaningful_words, data = generate_words(root_words, suffixes)
```

```python
print("Generated meaningful words:")
print(meaningful_words)
```

```python
df = pd.DataFrame(data, index=root_words)
print("Generated words from each root and suffix:")
df.head(50)
```

```python
meaningful_word_counts = df.apply(lambda row: sum(1 for value in row if value != False), axis=1)

plt.figure(figsize=(20, 10))
meaningful_word_counts.plot(kind='bar', color='skyblue')
plt.title("Number of Meaningful Words Formed from Each Root Word")
plt.xlabel("Root Words")
plt.ylabel("Count of Meaningful Words")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
